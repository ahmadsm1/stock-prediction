```python
import yfinance as yf
import pandas as pd
```

```python
sp500 = yf.Ticker("^GSPC")
sp500 = sp500.history(period="max")
```

```python
sp500
```

```python
sp500.index
```

```python
sp500.plot.line(y="Close", use_index=True)
```

```python
del sp500["Dividends"]
```

```python
del sp500["Stock Splits"]
```

```python
sp500["Tomorrow"] = sp500["Close"].shift(-1)
```

```python
sp500
```

```python
sp500["Pick"] = (sp500["Tomorrow"] > sp500["Close"]).astype(int)
```

```python
sp500
```

```python
sp500 = sp500.loc["1990-01-01":"2023-12-31"].copy()
```

```python
sp500
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model = RandomForestClassifier(n_estimators=100, min_samples_split=100, random_state=1)

train = sp500.iloc[:-100]
test = sp500.iloc[-100:]

predictors = ["Open", "High", "Low","Close","Volume"]
```

```python
model.fit(train[predictors], train["Pick"])
```

```python
from sklearn.metrics import precision_score
```

```python
predictions = model.predict(test[predictors])
```

```python
predictions = pd.Series(predictions, index=test.index)
```

```python
precision_score(test["Pick"], predictions)
```

```python
joined = pd.concat([test["Pick"], predictions], axis=1)
joined.plot()
```

```python
def predict(model, train, test, predictors):
    model.fit(train[predictors], train["Pick"])
    predictions = model.predict(test[predictors])
    predictions = pd.Series(predictions, index=test.index, name="Predictions")
    joined = pd.concat([test["Pick"], predictions], axis=1)
    return joined
```

```python
def backtest(data, model, predictors, start=2500, step=250):
    all_predictions = []

    for i in range(start, data.shape[0], step):
        train = data.iloc[0:i].copy()
        test = data.iloc[i:(i+step)].copy()
        predictions = predict(model, train, test, predictors)
        all_predictions.append(predictions)
    return pd.concat(all_predictions)
```

```python
predictions1 = backtest(sp500, model, predictors)
```

```python
predictions1["Predictions"].value_counts()
```

```python
precision_score(predictions1["Pick"], predictions1["Predictions"])
```

```python
predictions1["Pick"].value_counts() / predictions1.shape[0]
```

```python
horizons = [2,5,60,250,1000]
# rolling averages between last day, last week, last 3 months, last year, last 4 years
new_predictors = []

for horizon in horizons:
    rolling_averages = sp500.rolling(horizon).mean()

    ratio_column = f"Close_Ratio_{horizon}"
    sp500[ratio_column] = sp500["Close"] / rolling_averages["Close"]

    trend_column = f"Trend_{horizon}"
    sp500[trend_column] = sp500.shift(1).rolling(horizon).sum()["Pick"]

    new_predictors += [ratio_column, trend_column]
```

```python
sp500 = sp500.dropna()
```

```python
sp500
```

```python
model = RandomForestClassifier(n_estimators=200, min_samples_split=50, random_state=1)
```

```python
def predict(model, train, test, predictors):
    model.fit(train[predictors], train["Pick"])
    predictions = model.predict_proba(test[predictors])[:,1]
    predictions[predictions>=0.6]=1
    predictions[predictions<0.6]=0
    predictions = pd.Series(predictions, index=test.index, name="Predictions")
    joined = pd.concat([test["Pick"], predictions], axis=1)
    return joined
```

```python
predictions1 = backtest(sp500, model, new_predictors)
```

```python
predictions1["Predictions"].value_counts()
```

```python
precision_score(predictions1["Pick"], predictions1["Predictions"])
```

Our model was able to correctly predict the next value of the stock 57% of the time
